```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assume your dataframe is df

# --- Univariate Analysis ---

# Summary statistics for numeric columns
print(df.describe())

# Distribution plot of 'Total Net Worth'
plt.figure(figsize=(8,5))
sns.histplot(df['Total Net Worth'], bins=30, kde=True)
plt.title('Distribution of Total Net Worth')
plt.show()

# Countplot of 'Country' (categorical)
plt.figure(figsize=(18,24))
sns.countplot(y='Country', data=df, order=df['Country'].value_counts().index)
plt.title('Counts of Entries by Country')
plt.show()

# Boxplot of 'LastChange' to see spread and outliers
plt.figure(figsize=(8,5))
sns.boxplot(x=df['LastChange'])
plt.title('Boxplot of LastChange')
plt.show()
```

```python
# Scatter plot: Total Net Worth vs LastChange
plt.figure(figsize=(8,5))
sns.scatterplot(x='LastChange', y='Total Net Worth', data=df)
plt.title('Total Net Worth vs LastChange')
plt.show()

# Boxplot: Total Net Worth by Country (to see distribution by categories)

plt.figure(figsize=(12,18))
sns.boxplot(y='Country', x='Total Net Worth', data=df)
plt.title('Total Net Worth by Country')
plt.show()
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include='number')

# Plot correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
